```python
#Create Date Objects
#Here time is taken as zero by default , date is compulsory
y= dt.datetime(2020, 5, 17)
```

```python
#If you want date also give all the parameters
#Parameters for time and timezone (hour, minute, second, microsecond, tzone)
y= dt.datetime(2020, 7, 17,10,27,50)
```

```python
print(y)
```

```python
#strftime() method formats date objects into readable strings

print("Full Month :"+y.strftime("%B"))#Full Month; 
print("ShortMonth :"+y.strftime("%b"))#ShortMonth;
print("Month as a number :"+y.strftime("%m"))#Month as a number
print("short weekday :"+y.strftime("%a"))# Gives short weekday;
print("full weekday :"+y.strftime("%A"))# Gives full weekday;
print("Weekday as a number :"+y.strftime("%w"))#Weekday as a number 0-6, 0 is Sunday
print("Day of Month :"+y.strftime("%d"))# Day of Month;
```

```python
print("Year short :"+y.strftime("%y"))
print("Year long :"+y.strftime("%Y"))
```

```python
print("24 hours format :"+y.strftime("%H"))
print("12 hours format :"+y.strftime("%I"))
```

```python
print("minutes :"+y.strftime("%M"))
print("seconds :"+y.strftime("%S"))
print("local version :"+y.strftime("%x"))
```
